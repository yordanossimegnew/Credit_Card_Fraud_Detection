# Exploratory Data Analysis

### Importing necessary libraries and Loading the Cleaned Dataset

```python
# importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",None)
```

```python
# Loading the Cleaned Dataset
df = pd.read_csv("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\data\\processed\\cleaned_data.csv",)
df
```

```python
df.describe().round(2)
```

#### 1. Univariate Analysis

```python
# Univariate Analysis on Class
plt.figure(figsize = (6,4))
sns.countplot(data = df, x = "Class")
plt.title("Count Plot of Class")
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\reports\\figures\\1.0 Count Plot of Class.jpg")
plt.show()
```

From the above plot chart we can conclude that:
1. The number of records which are classified as frauds are almost negligable compared to those records classified as not fraud.

```python
# Univariate Analysis on Time
plt.figure(figsize = (10,4))
sns.distplot(df.Time)
plt.title("Distribution Plot of Time")
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\reports\\figures\\2.0 Distribution Plot of Time.jpg")
plt.show()
```

```python
# Univariate Analysis on Amount
plt.figure(figsize = (12,3))
sns.distplot(df.Amount)
plt.title("Distribution plot of Amount")
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\reports\\figures\\3.0 Distribution plot of Amount.jpg")
plt.show()
```

#### 2. Bivariate Analysis

```python
# Bivariate analysis on Amount vs Class
plt.figure(figsize= (6,3))
sns.boxenplot(data = df, x="Class",y = "Amount")
plt.title("Amount vs Class")
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\reports\\figures\\4.0 Amount vs Class.jpg")
plt.show()
```

```python
# Amount VS Class
plt.figure(figsize=(6,4))
sns.barplot(data = df, x="Class",y = "Amount")
plt.title("Amount Vs Class")
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\4.0 Fraud_Detection\\reports\\figures\\5.0 Amount Vs Class.jpg")
plt.show()
```
