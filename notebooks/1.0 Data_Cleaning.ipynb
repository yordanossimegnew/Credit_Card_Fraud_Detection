## 3 Data Cleaning

##### 3.1 Standardizing Column Names

```python
# defining a function to standardize column names
def column_maker(column_name):
    return column_name.str.title()
```

```python
# applying column_maker function
df.columns = column_maker(df.columns)
```

```python
# checking our column names
df.columns
```

##### 3.2 Removing Duplicated Records (if any)

```python
# first let's check the existance of duplicated records
df.duplicated().any()
```

the above result tells us the existance of duplicated records in our dataframe

```python
# let's see how many records are duplicated
df.duplicated().sum()
```

we have 1081 duplicated results in our data frame

```python
# now let's remove the duplicated records from our dataframe
df.drop_duplicates(inplace=True)
```

```python
# now let's check the shape of our new dataframe
df.shape
```

```python
# since we have removed records from our dataframe, we have to fix the index
df.reset_index()
```

```python
# now let's remove the index column from our dataframe
df.drop("index", axis = 1, inplace = True)
```
